```python
import matplotlib.pyplot as plt

import numpy as np
import torch
from torch.utils.data import DataLoader

from data.dataset import SegmentDataset
from data.transforms import transform
from data.collate import collate
# from utils.
```

```python
train_annot = '../assets/train_sample.txt'
batch_size = 1
```

```python
train_set = SegmentDataset(annot=train_annot, transform=transform, dim=(2048, 2048))
train_loader = DataLoader(
    train_set,
    batch_size=batch_size,
    num_workers=0,
    collate_fn=collate,
)
```

```python
for _, _, imgs, masks in train_loader:
    n = len(masks)
    mask = masks[0]
    plt.imshow(mask)
    plt.show()
    
    print(np.count_nonzero(mask==0))
    print(np.count_nonzero(mask==1))
    print(np.count_nonzero(mask==2))
    break
```

```python
for _, _, imgs, masks in train_loader:
    n = len(masks)
    mask = masks[0]
    
    mask = mask[250:500, 250:500]
    plt.imshow(mask)
    plt.show()
    
    print(np.count_nonzero(mask==0)) # background
    print(np.count_nonzero(mask==1)) # house
    print(np.count_nonzero(mask==2)) # road
    break
```
